# Train a Model on MNIST with FiftyOne and Torch

```python
import fiftyone as fo
import fiftyone.zoo as foz
import fiftyone.utils.random as four
import torch
```

```python
coco = foz.load_zoo_dataset("coco-2017")
coco.persistent = True
```

Launch fiftyone app

```python
session = fo.launch_app(coco, auto=False)
```

```python
coco.untag_samples(['train', 'validation'])

not_test = coco.match_tags('test', bool=False)
four.random_split(not_test, {'train' : 0.9, 'validation' : 0.1})
print(coco.count_sample_tags())
```
